```python
import cv2 as cv
import sys
import matplotlib.pyplot as plt

img = cv.imread('soccer.jpg')

if img is None:
    print('Image load failed')
    sys.exit()
    
h1 = cv.calcHist([img], [2], None, [256], [0, 256])  # (image, channel, mask, histSize, ranges)
plt.plot(h1, color='r')
plt.title('Red Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

```python
h2 = cv.calcHist([img], [1], None, [256], [0, 256])  # (image, channel, mask, histSize, ranges)
plt.plot(h2, color='g')
plt.title('Green Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

```python
h3 = cv.calcHist([img], [0], None, [256], [0, 256])  # (image, channel, mask, histSize, ranges)
plt.plot(h3, color='b')
plt.title('Green Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

```python
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
h4 = cv.calcHist([hsv], [0], None, [180], [0, 180])  # (image, channel, mask, histSize, ranges) Hue 범위는 0~180도!
plt.plot(h4, color='m')
plt.title('Hue Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

```python
lab = cv.cvtColor(img, cv.COLOR_BGR2Lab)
h5 = cv.calcHist([lab], [1], None, [256], [0, 256])  # (image, channel, mask, histSize, ranges)
plt.plot(h5, color='c')
plt.title('L Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```
