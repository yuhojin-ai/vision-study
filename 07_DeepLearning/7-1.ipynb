```python
import tensorflow as tf
import tensorflow.keras.datasets as ds  # keras   
import matplotlib.pyplot as plt          #  

# 2) MNIST     
(x_train, y_train), (x_test, y_test) = ds.mnist.load_data()  #   

print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
# : /   shape 

# 3) MNIST    
plt.figure(figsize=(24, 3))
plt.suptitle('MNIST', fontsize=30)

for i in range(10):
    plt.subplot(1, 10, i+1)  # 1 10  i+1
    plt.imshow(x_train[i], cmap='gray')  #   
    plt.xticks([]); plt.yticks([])       #   
    plt.title(str(y_train[i]), fontsize=30)  #   
    
# 4) CIFAR-10     
(x_train, y_train), (x_test, y_test) = ds.cifar10.load_data()  #   32x32

print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
# : /   shape 

# CIFAR-10   
class_names = [
    'airplane', 'car', 'bird', 'cat', 'deer',
    'dog', 'frog', 'horse', 'ship', 'truck'
]

# 5) CIFAR-10    
plt.figure(figsize=(24, 3))
plt.suptitle('CIFAR-10', fontsize=30)

for i in range(10):
    plt.subplot(1, 10, i+1)   # 1 10  i+1
    plt.imshow(x_train[i])    #   
    plt.xticks([]); plt.yticks([])  #   
    plt.title(class_names[y_train[i, 0]], fontsize=30)  #   
```
