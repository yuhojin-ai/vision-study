```python
import tensorflow as tf
import tensorflow.keras.datasets as ds  # keras에 내장된 데이터셋 모듈
import matplotlib.pyplot as plt          # 시각화 라이브러리

# 2) MNIST 데이터셋 로드 및 정보 출력
(x_train, y_train), (x_test, y_test) = ds.mnist.load_data()  # 손글씨 숫자 이미지

print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
# 출력: 학습/테스트 이미지와 레이블의 shape 확인

# 3) MNIST 데이터셋 일부 샘플 시각화
plt.figure(figsize=(24, 3))
plt.suptitle('MNIST', fontsize=30)

for i in range(10):
    plt.subplot(1, 10, i+1)  # 1행 10열 중 i+1번째
    plt.imshow(x_train[i], cmap='gray')  # 흑백 이미지 출력
    plt.xticks([]); plt.yticks([])       # 축 눈금 제거
    plt.title(str(y_train[i]), fontsize=30)  # 실제 레이블 표시
    
# 4) CIFAR-10 데이터셋 로드 및 정보 출력
(x_train, y_train), (x_test, y_test) = ds.cifar10.load_data()  # 컬러 이미지 32x32

print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
# 출력: 학습/테스트 이미지와 레이블의 shape 확인

# CIFAR-10 클래스 이름 정의
class_names = [
    'airplane', 'car', 'bird', 'cat', 'deer',
    'dog', 'frog', 'horse', 'ship', 'truck'
]

# 5) CIFAR-10 데이터셋 일부 샘플 시각화
plt.figure(figsize=(24, 3))
plt.suptitle('CIFAR-10', fontsize=30)

for i in range(10):
    plt.subplot(1, 10, i+1)   # 1행 10열 중 i+1번째
    plt.imshow(x_train[i])    # 컬러 이미지 출력
    plt.xticks([]); plt.yticks([])  # 축 눈금 제거
    plt.title(class_names[y_train[i, 0]], fontsize=30)  # 실제 클래스명 표시
```
