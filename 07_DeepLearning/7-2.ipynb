```python
import numpy as np
import tensorflow as tf
import tensorflow.keras.datasets as ds

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD

# [1] ë°ì´í„°ì…‹ ë¡œë“œ
# MNIST ì†ê¸€ì”¨ ë°ì´í„°ì…‹ ë¶ˆëŸ¬ì˜¤ê¸°
(x_train, y_train), (x_test, y_test) = ds.mnist.load_data()

# [2] ë°ì´í„° ì „ì²˜ë¦¬
# 28x28 ì´ë¯¸ì§€ë¥¼ 784 ë²¡í„°ë¡œ í‰íƒ„í™”
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)

# í”½ì…€ ê°’ì„ 0~1 ë²”ìœ„ë¡œ ì •ê·œí™”
x_train = x_train.astype(np.float32) / 255.0
x_test = x_test.astype(np.float32) / 255.0

# ë ˆì´ë¸”(ì •ë‹µ)ì„ one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

# [3] MLP ëª¨ë¸ êµ¬ì„±
mlp = Sequential()  # Sequential API ì‚¬ìš©

# ì€ë‹‰ì¸µ (ì…ë ¥: 784, ì¶œë ¥: 512), í™œì„± í•¨ìˆ˜: tanh
mlp.add(Dense(units=512, activation='tanh', input_shape=(784,)))

# ì¶œë ¥ì¸µ (í´ë˜ìŠ¤: 10ê°œ), í™œì„± í•¨ìˆ˜: softmax
mlp.add(Dense(units=10, activation='softmax'))

# [4] ëª¨ë¸ ì»´íŒŒì¼
# ì†ì‹¤ í•¨ìˆ˜: MSE (ë³´í†µ ë‹¤ì¤‘ë¶„ë¥˜ëŠ” categorical_crossentropy ì‚¬ìš© ê¶Œì¥)
# ìµœì í™” ì•Œê³ ë¦¬ì¦˜: SGD(í™•ë¥ ì  ê²½ì‚¬í•˜ê°•ë²•)
# í‰ê°€ ì§€í‘œ: ì •í™•ë„
mlp.compile(loss='MSE',
            optimizer=SGD(learning_rate=0.01),
            metrics=['accuracy'])

# [5] í•™ìŠµ ì‹¤í–‰
# ë°°ì¹˜ ì‚¬ì´ì¦ˆ: 128, ì—í­: 50íšŒ, ê²€ì¦ ë°ì´í„°: í…ŒìŠ¤íŠ¸ì…‹ ì‚¬ìš©
mlp.fit(x_train, y_train,
        batch_size=128,
        epochs=50,
        validation_data=(x_test, y_test),
        verbose=2)

# [6] ëª¨ë¸ í‰ê°€
# í…ŒìŠ¤íŠ¸ì…‹ì— ëŒ€í•œ ì •í™•ë„ ì¶œë ¥
res = mlp.evaluate(x_test, y_test, verbose=0)
print('ì •í™•ë¥  =', res[1] * 100)
```
